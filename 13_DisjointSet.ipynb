# DisjointSet

```python
class DisjointSet:
  def __init__(self, n):
    self.S = [-1]*n
  
  def find(self, a):
    if self.S[a] < 0:
      return a
    parent = self.find(self.S[a])
    self.S[a] = parent
    return parent

  def sameset(self, a, b):
    return self.find(a) == self.find(b)

  def union(self, a, b):
    if self.sameset(a, b):
      return
    rx = self.find(a)
    ry = self.find(b)
    
    if self.S[rx] < self.S[ry]:
      self.S[rx] += self.S[ry]
      self.S[ry] = rx
    else:
      self.S[ry] += self.S[rx]
      self.S[rx] = ry
```

```python
import graphviz as gv

def show(ds):
  G = gv.Digraph("happiness")
  G.graph_attr["rankdir"] = "BT"
  s = ds.S
  n = len(s)

  for u in range(n):
    G.node(str(u))

  for u, v in enumerate(s):
    if v >= 0:
      G.edge(str(u), str(v))
  return G
```

```python
dset = DisjointSet(10)
show(dset)
```
